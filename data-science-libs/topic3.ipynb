# Задание 2

```python
t = np.linspace(0, 10, num = 51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f, color = 'green')
plt.title("График f(t)")
plt.xlabel("Значения t")
plt.ylabel("Значения f")
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

# Задание 3

```python
x = np.linspace(-3, 3, num = 51)
x
```

```python
y1 = x**2
y1
```

```python
y2 = 2*x + 0.5
y2
```

```python
y3 = -3*x - 1.5
y3
```

```python
y4 = np.sin(x)
y4
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title("График y1")
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title("График y2")

ax3.plot(x, y3)
ax3.set_title("График y3")

ax4.plot(x, y4)
ax4.set_title("График y4")

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

plt.show()
```

Нашел стандартный способ вывода плотности распределения через `kind="kde"`. 
Как сделать то же самое через гистограммы средствами matplotlib - не ясно.

```python
not_fraud = df[df["Class"] == 0]
fraud = df[df["Class"] == 1]

not_fraud_kde = not_fraud["V1"].plot.kde(bw_method=3, color="gray", label="Class 0")
not_fraud_kde.legend()
fraud_kde = fraud["V1"].plot.kde(bw_method=3, color="red", label="Class 1")
legend = fraud_kde.legend()
```
