# Задание 1

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston

boston = load_boston()
print(boston["DESCR"])
```

```python
data = boston["data"]
feature_names = boston["feature_names"]
target = boston["target"]
```

```python
X = pd.DataFrame(data, columns=feature_names)
X
```

```python
y = pd.DataFrame(target, columns=["price"])
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr = lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})
check_test
```

```python
from sklearn.metrics import r2_score

r2_score(check_test["y_test"], check_test["y_pred"])
```

# Задание 2

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

```python
model = model.fit(X_train, y_train.values[:, 0])
```

```python
y_pred = model.predict(X_test)
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})
r2_score(check_test["y_test"], check_test["y_pred"])
```

При использовании RandomForestRegressor коэффицент детерминации ближе к единице, чем при использовании LinearRegression. Это говорит о том, что модель на основе RandomForestRegressor работает лучше.

# Задание 4

```python
df = pd.read_csv("creditcard.csv")
```

```python
df["Class"].value_counts(normalize=True)
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
```

```python
df.head(10)
```

```python
X = df.drop(['Class'], axis=1)
```

```python
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

parameters = {
    'n_estimators': [10, 15],
    'max_features': np.arange(3, 5),
    'max_depth': np.arange(4, 7)
}

clf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3
)
```

```python
clf = clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
y_pred_proba = clf.predict_proba(X_test)
y_pred_proba = y_pred_proba[:, 1]
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test, y_pred_proba)
```
